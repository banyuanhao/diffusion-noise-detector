```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

values = [0.1777, 0.1686, 0.1636, 0.1615, 0.1565, 0.1529, 0.1522, 0.1540, 0.1493, 0.14666]
groups = list(range(1, 11))

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Create orange bar chart
bars = ax.bar(groups, values, color='orange', alpha=0.7, width=0.7, edgecolor='grey')

# Smooth line plot over the bars
ax.plot(list(range(1,11,1)), values, color='darkblue', linewidth=2, alpha=0.8)

shifts_x = [0,0.3,0.1,0,0.2,0,0,0,0,0]
shifts_y = [0,0.001,0,0,0,0.001,0,0,0,0]
# Add data labels to each bar
for i,bar in enumerate(bars):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2+shifts_x[i], yval+0.002+shifts_y[i], round(yval, 4), va='bottom', ha='center', fontsize=13, color='black')

# Add grid, labels, and legend with font size adjustments
ax.set_ylabel('Energy Distance', fontsize=14)
ax.set_xlabel('Group Number', fontsize=14)
ax.set_xticks(groups)  # Set x-ticks to be group numbers
ax.grid(True, which='major', linestyle='--', linewidth='0.5', color='gray')

# # Set the background color
# fig.patch.set_facecolor('white')
# ax.set_facecolor('whitesmoke')

# Adjust y-axis to not start from zero
ax.set_ylim(min(values) - 0.01, max(values) + 0.005)  # Adjust the range as necessary
ax.tick_params(axis='both', which='major', labelsize=14)

# Save and show the figure
plt.tight_layout()
plt.savefig('/home/banyh2000/odfn/wrapup_data/distance_enhanced.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Change directory appropriately using os.path
import os
base_path = '/home/banyh2000/odfn'
os.chdir(base_path)

from scripts.utils.utils_odfn import variance_5_class

# Ensure data is sorted
variance_5_class.sort()

# Calculate mean entropy for each 2000 segment
entropy = [np.mean(variance_5_class[i:i+2000]) for i in range(0, 20000, 2000)]

# Energy distance values for each entropy segment
values = [0.1777, 0.1686, 0.1636, 0.1615, 0.1565, 0.1529, 0.1522, 0.1540, 0.1493, 0.14666]

# Plotting the data
plt.plot(entropy, values, label='Trigger Patch Group')

# Adding a baseline for control group comparison
base = 0.1331
plt.axhline(y=base, color='r', linestyle='-', label='Control Group')

# Adding grid, labels, and legend with larger font
plt.grid(True)
plt.ylabel('Energy Distance', fontsize=14)
plt.xlabel('Mean Trigger Entropy', fontsize=14)
plt.legend(prop={'size': 14})

# Save plot to file
plot_path = os.path.join(base_path, 'wrapup_data', 'distance.png')
plt.savefig(plot_path)
plt.show()  # Optionally display the plot as well
```
