```python
import numpy as np
import json
import matplotlib.pyplot as plt
import os
os.chdir('..')
from scripts.utils.utils_odfn import variance_5_class_index_sorted, seeds_plus

spilt = 'train'
mode = '1_category_5_class_npy'
name = spilt + '_for_' + mode + '.json'

base_path = '/nfs/data/yuanhaoban/ODFN/version_2/'
path = base_path + spilt + '/annotations/' + name
data = json.load(open(path, 'r'))
images = data['images']
annotations = data['annotations']

image_id = [variance_5_class_index_sorted[1210], variance_5_class_index_sorted[18900]]

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
fig.tight_layout()
for i in range(2):
    place_holder = np.zeros((64,64),dtype=np.float32)
    for ann in annotations:
        if ann['image_id'] == image_id[i]:
            print(ann['bbox'])
            bbox = ann['bbox']
            bbox = [int(i) for i in bbox]
            place_holder[bbox[1]:bbox[3],bbox[0]:bbox[2]] += 1
        
    place_holder = place_holder / np.max(place_holder)
    axs[i].imshow(place_holder)
    
# fig.show()
fig.savefig('/home/banyh2000/odfn/wrapup_data/heatmap.png')
```
